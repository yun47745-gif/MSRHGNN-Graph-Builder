```python
import os
import torch
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
from torch_geometric.data import HeteroData
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# =====================================================================
# 🛠️ 路径配置 (补回映射字典)
# =====================================================================
DATA_DIR = "/hy-tmp/else/data/data"
OUT_PT_PATH = "/hy-tmp/else/processed/final_hetero_data_raw.pt"

FILE_CTD_DISEASE = os.path.join(DATA_DIR, "CTD_diseases.csv")
FILE_CTD_CHEM_DIS = os.path.join(DATA_DIR, "CTD_chemicals_diseases.csv")
FILE_CTD_GENE_DIS = os.path.join(DATA_DIR, "CTD_curated_genes_diseases.csv.gz")
FILE_DRUGBANK_XML = os.path.join(DATA_DIR, "full database.xml")
FILE_ID_MAPPING = os.path.join(DATA_DIR, "id_mapping.csv") # 必须用到！UniProt -> Entrez
FILE_STRING_LINKS = os.path.join(DATA_DIR, "9606.protein.links.v12.0.txt.gz")
FILE_STRING_ALIASES = os.path.join(DATA_DIR, "9606.protein.aliases.v12.0.txt.gz")
DISEASE_KEYWORDS = [
    "Alzheimer", "Parkinson", "Huntington", "ALS",
    "schizophrenia", "depression", "bipolar",
    "dementia", "epilepsy", "autism"
]

# =====================================================================
# 🧠 通用安全读取函数（你的杰作，原封不动保留）
# =====================================================================
def read_ctd_safe(file_path, columns):
    df = pd.read_csv(file_path, dtype=str, header=None, comment='#', sep=None, engine='python')
    
    # 3. 动态对齐列名，防止文件实际列数少于我们提供的列名列表
    df.columns = columns[:df.shape[1]]
    
    # 如果只有一列 → 说明是 "\t"
    if df.shape[1] == 1:
        df = df.iloc[:, 0].str.split(r'\\t', expand=True)
    df.columns = columns[:df.shape[1]]
    # 去掉注释行
    df = df[~df.iloc[:, 0].str.startswith('#', na=False)]
    return df.fillna('')

# =====================================================================
# 🚀 主函数
# =====================================================================
def build_graph_from_raw():
    print("🚀 构建异构图（融合优化版）...")
    data = HeteroData()

    # =====================================================================
    # 🧬 1. 疾病 & 映射字典准备
    # =====================================================================
    print("\n[1/6] 读取疾病与映射字典...")
    cols_dis = ['DiseaseName', 'DiseaseID', 'AltDiseaseIDs', 'Definition',
                'ParentIDs', 'TreeNumbers', 'ParentTreeNumbers', 'Synonyms', 'SlimMappings']
    df_dis_all = read_ctd_safe(FILE_CTD_DISEASE, cols_dis)
    
    mask = df_dis_all['DiseaseName'].str.contains('|'.join(DISEASE_KEYWORDS), case=False, na=False)
    df_dis = df_dis_all[mask].reset_index(drop=True)

    disease_ids = df_dis['DiseaseID'].tolist()
    disease_id_to_idx = {d: i for i, d in enumerate(disease_ids)}
    data['disease'].node_id = disease_ids
    print(f"✅ 核心疾病数: {len(disease_ids)}")

    # 载入 UniProt -> Entrez 字典
    df_map = pd.read_csv(FILE_ID_MAPPING, sep=None, engine='python', dtype=str).dropna()
    df_map.columns = df_map.columns.str.strip()
    uniprot_to_entrez = dict(zip(df_map['From'].str.strip(), df_map['To'].str.strip()))

    # =====================================================================
    # 🎯 2. DrugBank XML 解析 (提取映射桥梁与靶点)
    # =====================================================================
    print("\n[2/6] 解析 DrugBank 建立 ID 桥梁与靶点池...")
    ns = '{http://www.drugbank.ca}'
    cas_to_db, mesh_to_db, db_to_uniprots = {}, {}, {}

    context = ET.iterparse(FILE_DRUGBANK_XML, events=('end',))
    for event, elem in context:
        if elem.tag == f'{ns}drug':
            id_elem = elem.find(f'{ns}drugbank-id[@primary="true"]')
            if id_elem is None:
                elem.clear(); continue
            db_id = id_elem.text

            # 提取 CAS 和 MeSH，用于后续桥接 CTD
            cas_elem = elem.find(f'{ns}cas-number')
            if cas_elem is not None and cas_elem.text: cas_to_db[cas_elem.text.strip()] = db_id
            ext_ids = elem.find(f'{ns}external-identifiers')
            if ext_ids is not None:
                for ext in ext_ids.findall(f'{ns}external-identifier'):
                    if ext.find(f'{ns}resource').text == 'MeSH':
                        mesh_to_db[ext.find(f'{ns}identifier').text.strip()] = db_id

            # 提取靶点 UniProt
            uniprots = []
            targets = elem.find(f'{ns}targets')
            if targets is not None:
                for t in targets.findall(f'{ns}target'):
                    poly = t.find(f'{ns}polypeptide')
                    if poly is not None and poly.get('id'):
                        uniprots.append(poly.get('id'))
            if uniprots: db_to_uniprots[db_id] = uniprots
            
            elem.clear()

    print(f"✅ 提取到 {len(cas_to_db)}个CAS, {len(mesh_to_db)}个MeSH, {len(db_to_uniprots)}个药物带靶点")

    # =====================================================================
    # 💊 3. 物理边：药物-疾病 & 疾病-基因
    # =====================================================================
    print("\n[3/6] 构建物理边 (关联与治疗)...")
    
    # --- 疾病-基因 ---
    cols_gene_dis = ['GeneSymbol', 'GeneID', 'DiseaseName', 'DiseaseID',
                     'DirectEvidence', 'InferenceChemicalName', 'InferenceScore', 'OmimIDs', 'PubMedIDs']
    df_gene_dis = read_ctd_safe(FILE_CTD_GENE_DIS, cols_gene_dis)
    df_gene_dis = df_gene_dis[df_gene_dis['DiseaseID'].isin(disease_ids)]
    df_gene_dis = df_gene_dis[df_gene_dis['DirectEvidence'] != '']

    gene_ids = df_gene_dis['GeneID'].unique().tolist()
    gene_id_to_idx = {g: i for i, g in enumerate(gene_ids)}
    data['gene'].node_id = gene_ids

    src_assoc, dst_assoc = [], []
    for _, row in df_gene_dis.iterrows():
        dis, g = row['DiseaseID'], row['GeneID']
        if dis in disease_id_to_idx and g in gene_id_to_idx:
            src_assoc.append(disease_id_to_idx[dis])
            dst_assoc.append(gene_id_to_idx[g])
            
    data['disease', 'associated_with', 'gene'].edge_index = torch.unique(torch.tensor([src_assoc, dst_assoc], dtype=torch.long), dim=1)
    print(f"✅ 基因数: {len(gene_ids)} | 关联边: {data['disease', 'associated_with', 'gene'].edge_index.shape[1]}")

    # --- 药物-疾病 ---
    cols_chem_dis = ['ChemicalName', 'ChemicalID', 'CasRN', 'DiseaseName',
                     'DiseaseID', 'DirectEvidence', 'InferenceGeneSymbol', 'InferenceScore', 'OmimIDs', 'PubMedIDs']
    df_chem_dis = read_ctd_safe(FILE_CTD_CHEM_DIS, cols_chem_dis)
    df_chem_dis = df_chem_dis[df_chem_dis['DiseaseID'].isin(disease_ids)]
    df_chem_dis = df_chem_dis[df_chem_dis['DirectEvidence'].str.contains('therapeutic', case=False, na=False)]

    drug_id_set = set()
    src_treats_tmp, dst_treats_tmp = [], []
    for _, row in df_chem_dis.iterrows():
        dis_id = row['DiseaseID']
        mesh_raw = row['ChemicalID'].replace('MESH:', '').strip()
        cas_raw = row['CasRN'].strip()
        
        # 桥接获取真正的 DrugBank ID
        db_id = cas_to_db.get(cas_raw) or mesh_to_db.get(mesh_raw)
        if db_id:
            drug_id_set.add(db_id)
            src_treats_tmp.append(db_id)
            dst_treats_tmp.append(disease_id_to_idx[dis_id])

    drug_ids = list(drug_id_set)
    drug_id_to_idx = {d: i for i, d in enumerate(drug_ids)}
    data['drug'].node_id = drug_ids
    
    src_treats = [drug_id_to_idx[d] for d in src_treats_tmp]
    data['drug', 'treats', 'disease'].edge_index = torch.unique(torch.tensor([src_treats, dst_treats_tmp], dtype=torch.long), dim=1)
    print(f"✅ 药物数: {len(drug_ids)} | 治疗边: {data['drug', 'treats', 'disease'].edge_index.shape[1]}")

    # =====================================================================
    # 🎯 4. 高阶边：药物-靶点 (UniProt -> Entrez 映射)
    # =====================================================================
    print("\n[4/6] 构建药物-靶点高阶边...")
    src_targets, dst_targets = [], []
    for db_id in drug_ids:
        if db_id in db_to_uniprots:
            for up_id in db_to_uniprots[db_id]:
                entrez_id = uniprot_to_entrez.get(up_id)
                if entrez_id and entrez_id in gene_id_to_idx:
                    src_targets.append(drug_id_to_idx[db_id])
                    dst_targets.append(gene_id_to_idx[entrez_id])
                    
    if src_targets:
        edge = torch.unique(torch.tensor([src_targets, dst_targets], dtype=torch.long), dim=1)
        data['drug', 'targets', 'gene'].edge_index = edge
        print(f"✅ 成功打通 drug-target 元路径！边数: {edge.shape[1]}")
    else:
        print("⚠️ 未匹配到 drug-target 边。")

    # =====================================================================
    # 🌐 5. 疾病相似度三网络 (H, G, T)
    # =====================================================================
    print("\n[5/6] 构建疾病相似度三网络 (MSRHGNN 核心)...")
    num_diseases = len(disease_ids)
    
    # --- sim_h (树层级) ---
    # --- sim_h (本体层级：基于 ParentIDs 修复版) ---
    sim_h_mat = np.zeros((num_diseases, num_diseases))
    # 核心修复：改用 ParentIDs 寻找共同祖先，抛弃容易缺失的 TreeNumbers
    parent_list = [set(str(p).split('|')) if p else set() for p in df_dis['ParentIDs']]
    
    for i in range(num_diseases):
        for j in range(i+1, num_diseases):
            p_i, p_j = parent_list[i], parent_list[j]
            if p_i and p_j: 
                sim_h_mat[i, j] = sim_h_mat[j, i] = len(p_i & p_j) / len(p_i | p_j)
                
    idx_h = np.nonzero(sim_h_mat)
    data['disease', 'sim_h', 'disease'].edge_index = torch.tensor(np.stack(idx_h), dtype=torch.long)
    data['disease', 'sim_h', 'disease'].edge_attr = torch.tensor(sim_h_mat[idx_h], dtype=torch.float).unsqueeze(1)
    
    # 2. 解析 Links 构建 PPI 图 (保留 score > 400 的中高置信度边)
    df_links = pd.read_csv(FILE_STRING_LINKS, sep=' ')
    df_links = df_links[df_links['combined_score'] > 400]
    
    ppi_graph = {}
    for _, row in df_links.iterrows():
        e1 = ensp_to_entrez.get(row['protein1'])
        e2 = ensp_to_entrez.get(row['protein2'])
        if e1 and e2 and e1 in gene_id_to_idx and e2 in gene_id_to_idx:
            idx1, idx2 = gene_id_to_idx[e1], gene_id_to_idx[e2]
            if idx1 not in ppi_graph: ppi_graph[idx1] = set()
            if idx2 not in ppi_graph: ppi_graph[idx2] = set()
            ppi_graph[idx1].add(idx2)
            ppi_graph[idx2].add(idx1)
            
    # 3. 基于 PPI 网络扩展疾病的关联基因集
    sim_g_mat = np.zeros((num_diseases, num_diseases))
    dis_to_genes = {i: set() for i in range(num_diseases)}
    for s, d in zip(src_assoc, dst_assoc): 
        dis_to_genes[s].add(d)
        
    dis_to_expanded_genes = {}
    for d_idx, g_set in dis_to_genes.items():
        expanded = set(g_set)
        for g in g_set:
            if g in ppi_graph:
                expanded.update(ppi_graph[g]) # 把 PPI 网络里的一阶邻居也加进来
        dis_to_expanded_genes[d_idx] = expanded
        
    # 4. 计算扩展后的 Jaccard 互作重合度
    for i in range(num_diseases):
        for j in range(i+1, num_diseases):
            g_i, g_j = dis_to_expanded_genes[i], dis_to_expanded_genes[j]
            if g_i and g_j:
                u_len = len(g_i | g_j)
                if u_len > 0: 
                    sim_g_mat[i, j] = sim_g_mat[j, i] = len(g_i & g_j) / u_len
                    
    idx_g = np.nonzero(sim_g_mat)
    data['disease', 'sim_g', 'disease'].edge_index = torch.tensor(np.stack(idx_g), dtype=torch.long)
    data['disease', 'sim_g', 'disease'].edge_attr = torch.tensor(sim_g_mat[idx_g], dtype=torch.float).unsqueeze(1)
    print(f"  -> STRING PPI 强化完成！sim_g边: {len(idx_g[0])}")

    # --- sim_t (使用你优化的 KNN 算法) ---
    # --- sim_t (强化版：完美对齐文档的临床文本语义) ---
    print("  -> 提取 CTD 临床病理描述，构建 TF-IDF 语义网络...")
    
    # 核心修复：必须加入 'Definition' 列！这才是真正的病理描述 
    texts = (df_dis['DiseaseName'] + " " + df_dis['Definition'] + " " + df_dis['Synonyms']).tolist()
    
    # 利用TF-IDF 算法提取疾病描述的语义特征 
    # 建议加上 max_features 限制维度，防止生成的 data['disease'].x 维度过大导致 OOM
    tfidf = TfidfVectorizer(stop_words='english', max_features=1024)
    X = tfidf.fit_transform(texts)
    
    k = 5
    nbrs = NearestNeighbors(n_neighbors=min(k+1, X.shape[0]), metric='cosine').fit(X)
    dist, idx = nbrs.kneighbors(X)

    src_t, dst_t, val_t = [], [], []
    for i in range(len(idx)):
        for j, d in zip(idx[i][1:], dist[i][1:]):
            src_t.append(i)
            dst_t.append(j)
            val_t.append(1 - d) # cosine distance 转为 similarity

    data['disease', 'sim_t', 'disease'].edge_index = torch.tensor([src_t, dst_t], dtype=torch.long)
    data['disease', 'sim_t', 'disease'].edge_attr = torch.tensor(val_t, dtype=torch.float).unsqueeze(1)
    
    # 将真正饱含临床语义的特征注入疾病节点
    data['disease'].x = torch.tensor(X.toarray(), dtype=torch.float)
    print(f"✅ sim_t 文本语义网络就绪！边数: {len(src_t)}")
    # =====================================================================
    # 🎯 6. 特征与保存
    # =====================================================================
    print("\n[6/6] 初始化特征并保存...")
    data['drug'].x = torch.randn((len(drug_ids), 128))
    # 结合你的思路，将 NLP 特征直接注入疾病节点初始特征
    data['disease'].x = torch.tensor(X.toarray(), dtype=torch.float)
    data['gene'].x = torch.randn((len(gene_ids), 128))

    os.makedirs(os.path.dirname(OUT_PT_PATH), exist_ok=True)
    torch.save(data, OUT_PT_PATH)

    print("\n🎉 构建完成！这套组合拳彻底消灭了图断裂。")
    print(data)

if __name__ == "__main__":
    build_graph_from_raw()
```

```python
import torch

PT_PATH = "/hy-tmp/else/processed/final_hetero_data_raw.pt"

def inspect_hetero_data(pt_path):
    print("🚀 加载数据...")
    data = torch.load(pt_path)

    print("\n" + "="*60)
    print("📊 基本信息")
    print("="*60)

    print(f"节点类型: {list(data.node_types)}")
    print(f"边类型: {list(data.edge_types)}")

    # ============================================================
    # 🧬 节点信息
    # ============================================================
    print("\n" + "="*60)
    print("🧬 节点详情")
    print("="*60)

    total_nodes = 0

    for ntype in data.node_types:
        node_data = data[ntype]

        num_nodes = node_data.num_nodes
        total_nodes += num_nodes

        print(f"\n🔹 节点类型: {ntype}")
        print(f"  数量: {num_nodes}")

        # 特征
        if hasattr(node_data, 'x'):
            print(f"  特征维度: {tuple(node_data.x.shape)}")
        else:
            print("  ⚠️ 无特征")

        # node_id
        if hasattr(node_data, 'node_id'):
            ids = node_data.node_id
            print(f"  ID示例: {ids[:5]}")
        else:
            print("  ⚠️ 无 node_id")

    print(f"\n🧮 总节点数: {total_nodes}")

    # ============================================================
    # 🔗 边信息
    # ============================================================
    print("\n" + "="*60)
    print("🔗 边详情")
    print("="*60)

    total_edges = 0

    for etype in data.edge_types:
        edge_data = data[etype]

        num_edges = edge_data.edge_index.shape[1]
        total_edges += num_edges

        print(f"\n🔸 边类型: {etype}")
        print(f"  边数量: {num_edges}")

        # 是否有权重
        if hasattr(edge_data, 'edge_attr'):
            print(f"  权重维度: {tuple(edge_data.edge_attr.shape)}")
        else:
            print("  无权重")

        # 示例边
        ei = edge_data.edge_index
        print("  示例边:")
        for i in range(min(5, num_edges)):
            print(f"    {ei[0, i]} -> {ei[1, i]}")

    print(f"\n🧮 总边数: {total_edges}")

    # ============================================================
    # 📈 密度 & 稀疏性
    # ============================================================
    print("\n" + "="*60)
    print("📈 图统计")
    print("="*60)

    for etype in data.edge_types:
        src_type, _, dst_type = etype
        num_src = data[src_type].num_nodes
        num_dst = data[dst_type].num_nodes
        num_edges = data[etype].edge_index.shape[1]

        density = num_edges / (num_src * num_dst)

        print(f"{etype}:")
        print(f"  密度: {density:.6e}")

    print("\n🎉 检查完成！")

# ============================================================
if __name__ == "__main__":
    inspect_hetero_data(PT_PATH)
```
